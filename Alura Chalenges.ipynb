# Importando Dados

```python
import pandas as pd
import matplotlib
%matplotlib inline
from matplotlib import pyplot as plt
from pandas.plotting import autocorrelation_plot
```

```python
dados_voz = pd.read_json('https://raw.githubusercontent.com/sthemonica/alura-voz/main/Dados/Telco-Customer-Churn.json',
                         )
dados_voz.head()
```

## Tratando dados

```python
df_extracao =  dados_voz.drop(['customerID', 'Churn'], axis=1)
```

```python
dados = dados_voz[['customerID', 'Churn']]
```

```python
for coluna in df_extracao:
    coluna_normalizada = pd.json_normalize(df_extracao[coluna])
    dados = pd.concat([dados, coluna_normalizada], axis=1)

dados.head()
```

**Entendendo as informações do conjunto**

```python
print(f'Quantidade de linhas e colunas {dados.shape}')
print(f'Quantidade de dados nulos {dados.isna().sum()}')
```

```python
dados.columns
```

```python
traducao_colunas =  {'customerID' : 'n_identificacao', 
                     'Churn' : 'deixou_empresa', 
                    'gender' : 'genero', 
                    'SeniorCitizen' : 'idoso', 
                    'Partner' : 'casado',
                    'Dependents' : 'dependentes', 
                    'tenure' : 'meses_contratados', 
                    'PhoneService' : 'assinatura_telefonica', 
                    'MultipleLines' : 'multiplas_linhas_tel',
                    'InternetService' : 'assinatura_internet', 
                    'OnlineSecurity' : 'assinatura_seguranca_online', 
                    'OnlineBackup' : 'assinatura_backup', 
                    'DeviceProtection' : 'proteca_dispositivo',
                    'TechSupport' : 'suporte_tecnico', 
                    'StreamingTV' : 'assinatura_cabo_tv', 
                    'StreamingMovies' : 'assinatura_stream', 
                    'Contract' : 'contrato',
                    'PaperlessBilling' : 'pagamento_online', 
                    'PaymentMethod' : 'tipo_fatura', 
                    'Charges.Monthly' : 'total_mensal',
                    'Charges.Total' : 'total_gasto'}
```

```python
dados = dados.rename(columns = traducao_colunas)
dados.sample(3)
```

### Definindo indice

```python
dados = dados.set_index(['n_identificacao'])
dados.sample(2)
```
